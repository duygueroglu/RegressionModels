```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
from sklearn.tree import DecisionTreeRegressor
```

```python
data = pd.read_csv("positions.csv")
```

```python
level = data.iloc[:,1:2].values.reshape(-1,1) #reshapes are for sklearn
salary = data.iloc[:,2].values.reshape(-1,1)
```

```python
regression = DecisionTreeRegressor()
regression.fit(level,salary)
print(regression.predict([[8.9]]))
```

```python
plt.scatter(level,salary,color="red")
x = np.arange(min(level),max(level),0.01).reshape(-1,1) #little range like 0.01
plt.plot(x,regression.predict(x),color="orange")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.title("Decision Tree Model")
plt.show()
```
